```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#Definition der Anfangswerte:
r = 20
t = 0
phi = 0
theta = math.pi / 2 
r_strich = 0
phi_strich = 0.036
t_strich = 1 / math.sqrt(1-2/r) * math.sqrt(1+r**2*phi_strich**2)
theta_strich = 0


dlamb = 0.001
listr = []
listt = []
lamblist = []
lamb=0
philist = []

while lamb < 150:
        listr.append(r)        
        listt.append(t)
        philist.append(phi)

        lamblist.append(lamb)
        lamb += dlamb

        dt = t_strich * dlamb
        dphi = phi_strich * dlamb
        dr = r_strich * dlamb
        dt_strich = 2*t_strich*r_strich / (-r**2 + 2*r) * dlamb
        dphi_strich = (-2 * r_strich * phi_strich / r) * dlamb
        dr_strich = ((-r+2)*t_strich**2/r**3 - r_strich**2/(r*(-r+2) -
(-r+2)*phi_strich**2) )* dlamb                                                


        t += dt
        phi += dphi
        r += dr
        t_strich += dt_strich
        phi_strich += dphi_strich
        r_strich += dr_strich



a=plt.figure(1)
plt.plot(lamblist,listt, linewidth=2, color='g')
plt.xlim([0, 79.5])
plt.ylim([0,150])
plt.xlabel('Eigenzeit')
plt.ylabel('Zeit')
a.savefig('Zeit.png')

b=plt.figure(2)
plt.plot(lamblist,listr,linewidth=2, color='r')
plt.xlim([0, 79.5])
plt.ylim([0, 100])
plt.xlabel('Eigenzeit')
plt.ylabel('Abstand r')
b.savefig('Abstand.png')

c=plt.figure(3)
plt.plot(lamblist,philist,linewidth=2, color='y')
plt.ylim([0, 100])
plt.xlim([0, 79.5])
plt.xlabel('Eigenzeit')
plt.ylabel('Phi')
c.savefig('Phi.png')
```
